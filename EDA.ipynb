```python
# function to count plot for categorical variables
def pointplot(var):

    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(15,8))
    
    sns.countplot(x=var, data=new0_app_data,ax=ax1)
    ax1.set_ylabel('Total Counts')
    ax1.set_title(f'Distribution of {var} for Non-Defaulters',fontsize=15)
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=30, ha="right")
    
    # Adding the normalized percentage for easier comparision between defaulter and non-defaulter
    for p in ax1.patches:
        ax1.annotate('{:.1f}%'.format((p.get_height()/len(new0_app_data))*100), (p.get_x()+0.1, p.get_height()+50))
        
    sns.countplot(x=var, data=new1_app_data,ax=ax2)
    ax2.set_ylabel('Total Counts')
    ax2.set_title(f'Distribution of {var} for Defaulters',fontsize=15)    
    ax2.set_xticklabels(ax2.get_xticklabels(), rotation=30, ha="right")
    
    # Adding the normalized percentage for easier comparision between defaulter and non-defaulter
    for p in ax2.patches:
        ax2.annotate('{:.1f}%'.format((p.get_height()/len(new1_app_data))*100), (p.get_x()+0.1, p.get_height()+50))
    
    plt.show()
```

```python
def plotunidist(var):

    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(15,5))
    
    sns.distplot(a=new0_app_data[var],ax=ax1)

    ax1.set_title(f'Distribution of {var} for Non-Defaulters',fontsize=15)
            
    sns.distplot(a=new1_app_data[var],ax=ax2)
    ax2.set_title(f'Distribution of {var} for Defaulters',fontsize=15)    
        
    plt.show()
```

```python
def plotbivar(var1,var2):

    plt.style.use('ggplot')
    sns.despine
    fig,(ax1,ax2) = plt.subplots(1,2,figsize=(20,8))
    
    sns.scatterplot(x=var1, y=var2,data=new0_app_data,ax=ax1)
    ax1.set_xlabel(var1)    
    ax1.set_ylabel(var2)
    ax1.set_title(f'{var1} vs {var2} for Non-Defaulters',fontsize=12)
    
    sns.scatterplot(x=var1, y=var2,data=new1_app_data,ax=ax2)
    ax2.set_xlabel(var1)    
    ax2.set_ylabel(var2)
    ax2.set_title(f'{var1} vs {var2} for Defaulters',fontsize=12)
            
    plt.show()
```

```python
def plot_by_cat_num(cat, num):

    plt.style.use('ggplot')
    sns.heatmap
    fig,ax = plt.subplots(1,1,figsize=(10,8))
    
    sns.boxenplot(x=cat,y = num, data=prev_data)
    ax.set_ylabel(f'{num}')
    ax.set_xlabel(f'{cat}')

    ax.set_title(f'{cat} Vs {num}',fontsize=15)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
     
    plt.show()
```

```python
plot_by_cat_num('NAME_CONTRACT_STATUS', 'AMT_ANNUITY')
```

```python
##Merging the files and analyzing the data
```

```python
NewLeftPrev = pd.merge(app_data, prev_data, how='left', on=['SK_ID_CURR'])
```

```python
NewLeftPrev.shape
```

```python
def plotuni_combined(Varx,Vary):
    # 100% bar chart
    plt.style.use('ggplot')
    sns.jointplot
    NewDat = NewLeftPrev.pivot_table(values='SK_ID_CURR', 
                      index=Varx,
                      columns=Vary,
                      aggfunc='count')
    NewDat=NewDat.div(NewDat.sum(axis=1),axis='rows')*100
    sns.set()
    NewDat.plot(kind='bar',stacked=True,figsize=(15,7))
    plt.title(f'Effect Of {Varx} on Loan Approval')
    plt.xlabel(f'{Varx}')
    plt.ylabel(f'{Vary}%')
    plt.show()
```

```python
plotuni_combined('FLAG_OWN_CAR','NAME_CONTRACT_STATUS')
```

```python
plotuni_combined('CODE_GENDER','NAME_CONTRACT_STATUS')
```

```python
plotuni_combined('TARGET','NAME_CONTRACT_STATUS')
```
